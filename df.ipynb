```python
#df_title 以title統計票房
files = glob('./test/new/20*.csv')

df_all = pd.concat(
    (pd.read_csv(file, usecols=
                 ['distributor','中文片名','銷售票數','銷售金額'], 
                 thousands=',',encoding='UTF-8') 
                 for file in files),ignore_index=True)

df_all = df_all.set_index('中文片名')
df_all = df_all.sort_index()

df_all_t = df_all.rename(columns={'銷售票數': 'Total 銷售票數', '銷售金額': 'Total 銷售金額'})
df_all_by_title = df_all_t.groupby('中文片名').sum()
df_all_by_title.sort_values('Total 銷售票數',ascending=False)
# df_all_by_title.to_csv("2018-2023 df_all_by_title.csv")
# df_all_by_title.to_excel("2018-2023 df_all_by_title.xlsx")
```

```python
#以出品公司統計cp

files = glob('./test/new/20*.csv')

df_all = pd.concat(
    (pd.read_csv(file, usecols=
                 ['distributor','中文片名','銷售票數','銷售金額'], 
                 thousands=',',encoding='UTF-8') 
                 for file in files),ignore_index=True)


df_all = df_all.set_index('distributor')
df_all = df_all.sort_index()

df_all_c = df_all.rename(columns={'銷售票數': 'Total 銷售票數', '銷售金額': 'Total 銷售金額'})
df_all_by_cp = df_all_c.groupby('distributor').sum()
df_all_by_cp.sort_values('Total 銷售票數',ascending=False)


# df_all_by_cp.to_csv("2018-2023 cp.csv")
# df_all_by_cp.to_excel("2018-2023 cp.xlsx")
```

```python
#以整體年度季度統計票房
files = glob('./test/new/20*.csv')

df_all = pd.concat(
    (pd.read_csv(file, usecols=
                 ['銷售票數','銷售金額','Date Range'], 
                 thousands=',',encoding='UTF-8') 
                 for file in files),ignore_index=True)

df_all['Date Range'] = pd.to_datetime(df_all['Date Range'])
df_all = df_all.set_index('Date Range')
df_all = df_all.sort_index()

df_all_q = df_all.resample('Q').sum()
df_all_q = df_all_q.rename(columns={'銷售票數': 'Total 銷售票數', '銷售金額': 'Total 銷售金額'})
df_all_by_quarter = df_all_q.groupby(df_all_q.index.to_period('Q')).sum()
df_all_by_quarter
# df_all_by_quarter.to_csv("2018-2023 by_quarter.csv")
# df_all_by_quarter.to_excel("2018-2023 by_quarter.xlsx")
```

```python
#以國家地區統計

files = glob('./test/new/20*.csv')

df_all = pd.concat(
    (pd.read_csv(file, usecols=
                 ['國別地區','銷售票數','銷售金額'], 
                 thousands=',',encoding='UTF-8') 
                 for file in files),ignore_index=True)


df_all = df_all.set_index('國別地區')
df_all = df_all.sort_index()

df_all_n = df_all.rename(columns={'銷售票數': 'Total 銷售票數', '銷售金額': 'Total 銷售金額'})
df_all_by_nation = df_all_n.groupby('國別地區').sum()


df_all_by_nation

# df_all_by_nation.to_csv("2018-2023 df_all_by_nation.csv")
# df_all_by_nation.to_excel("2018-2023 df_all_by_nation.xlsx")
```

```python
#以整體年度季度統計發行片量
files = glob('./test/new/20*.csv')

df_all = pd.concat(
    (pd.read_csv(file, usecols=
                 ['中文片名','Date Range'], 
                 thousands=',',encoding='UTF-8') 
                 for file in files),ignore_index=True)

df_all['Date Range'] = pd.to_datetime(df_all['Date Range'])
df_all = df_all.set_index('Date Range')
df_all = df_all.sort_index()

df_title_by_quarter = df_all.resample('Q').apply(lambda x: x['中文片名'].drop_duplicates().count())
df_title_by_quarter = df_title_by_quarter.to_frame('不重複的中文片名數量')  # list轉dataframe

df_title_by_quarter
# df_title_by_quarter.to_csv("2018-2023 df_title_by_quarter.csv")
# df_title_by_quarter.to_excel("2018-2023 df_title_by_quarter.xlsx")
```

```python
#以上映週數統計

files = glob('./test/new/20*.csv')

df_all = pd.concat(
    (pd.read_csv(file, usecols=
                 ['中文片名','Date Range'], 
                 thousands=',',encoding='UTF-8') 
                 for file in files),ignore_index=True)


df_all = df_all.set_index('中文片名')
df_all = df_all.sort_index()

df_all_by_times = df_all.groupby('中文片名').count()


df_all_by_times.sort_values('Date Range',ascending=False)

# df_all_by_times.to_csv("2018-2023 df_all_by_times.csv")
# df_all_by_times.to_excel("2018-2023 df_all_by_times.xlsx")
```

```python
#統計不同片商發片量
files = glob('./test/new/20*.csv')

df_all = pd.concat(
    (pd.read_csv(file, usecols=
                 ['出品','中文片名'], 
                 thousands=',',encoding='UTF-8') 
                 for file in files),ignore_index=True)

df_all_by_cps = df_all.drop_duplicates(subset=['出品', '中文片名']).groupby('出品')['中文片名'].count()
df_all_by_cps = pd.DataFrame(df_all_by_cps)
df_all_by_cps = df_all_by_cps.sort_values('中文片名',ascending=False)

df_all_by_cps

# df_all_by_cps.to_csv("2018-2023 df_all_by_cps.csv")
# df_all_by_cps.to_excel("2018-2023 df_all_by_cps.xlsx")
```
