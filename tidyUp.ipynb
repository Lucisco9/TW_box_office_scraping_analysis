```python
#測試檔案名稱重新命名

# 定義日期格式的正則表達式
date_regex = r"\d{4}年\d{1,2}月\d{1,2}日至?\d{0,4}年?\d{1,2}月\d{1,2}日?"

# 測試字串
test_str = "2019年1月29日至05月05日.csv"

# 使用 re.findall() 函數找出符合日期格式的字串
dates = re.findall(date_regex, test_str)

# 印出找到的日期
print(dates)
```

```python
##檔案名稱取出日期值並加入檔案欄位中
import csv
import os
import re
import datetime
source_path = "./test/"
dest_path = "./test/new/"
pattern = r"\d{4}年\d{1,2}月\d{1,2}日至?\d{0,4}年?\d{1,2}月\d{1,2}日?\.csv"

for filename in os.listdir(source_path):
    if re.match(pattern, filename):
        date_match = re.search(r"\d{4}年\d{1,2}月\d{1,2}日至?\d{0,4}年?\d{1,2}月\d{1,2}日?", filename)
        date_str = date_match.group()
        start_date, end_date = date_str.split("至")
        start_date = start_date.replace("年", "-").replace("月", "-").replace("日", "")
        n_start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d")
        #print(n_start_date)#印出來檢查一下是否正確
     
     
        with open(os.path.join(source_path, filename), "r", newline="", encoding="utf-8") as file:
           
            reader = csv.reader(file)
            rows = list(reader)
            
            # 把檔名中的日期放入週報表中，以便合併資料時可以辨識來自哪一週的報表
            rows[0].append("Date Range")
            for row in rows[1:]:
                row.append(f"{n_start_date:%Y/%m/%d}")

            # 把檔名改成只有日期
            new_filename = filename.replace(date_str, n_start_date.strftime("%Y-%m-%d"))
            #print(f"New file name: {new_filename}") #印出來檢查一下是否正確
        
    
        # 改名後另存新檔
        
            with open(os.path.join(dest_path, new_filename), "w", newline="", encoding="utf-8") as file:
                writer = csv.writer(file)
                writer.writerows(rows)
                print(f" {filename} 成功更名 {new_filename}")
                
```
